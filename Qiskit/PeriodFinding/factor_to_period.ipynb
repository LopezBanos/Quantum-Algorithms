```python
#=============================================================================#
#                            Importing packages                               #
#=============================================================================#
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
```

```python
#Find two prime numbers and take its product
p = 3
q = 5
N = p*q

print("The big prime number we need to factor is ", N)
a = 13
print("The coprime number of ", N, " is chosen to be ", a)
```

```python
#Find the period of a^r (mod 15):
r = np.arange(15)
y = np.array([])
for i in r:
    y = np.append(y,(a**i)%N)

print(y)
```

```python
#=============================================================================#
#                                Plotting                                     #
#=============================================================================#
plt.figure(figsize =(20,10))
plt.plot(r,y,color = "blue")
plt.hlines(1,4,8,color = "red")
plt.text(5.5,1.1,"Period",color = "red")
plt.title("From factoring to period finding")
plt.xlabel("Exponent: r")
plt.ylabel(r'$a^{r} (mod N)$')
plt.savefig('Factor_Period.png')
plt.show()
```
